```python
import torch

inputs = torch.tensor(
  [[0.43, 0.15, 0.89], # Your     (x^1)
   [0.55, 0.87, 0.66], # journey  (x^2)
   [0.57, 0.85, 0.64], # starts   (x^3)
   [0.22, 0.58, 0.33], # with     (x^4)
   [0.77, 0.25, 0.10], # one      (x^5)
   [0.05, 0.80, 0.55]] # step     (x^6)
)
```

```python
din = inputs.shape[1]
dout = 2
```

```python
Wq = torch.nn.Parameter(torch.rand(din, dout), requires_grad=False)
Wk = torch.nn.Parameter(torch.rand(din, dout), requires_grad=False)
Wv = torch.nn.Parameter(torch.rand(din, dout), requires_grad=False)
```

```python
Key = inputs @ Wk
Query = inputs @ Wk
Value = inputs @ Wv
```

```python
attention_score = Query @ Key.T
attention_weight = torch.softmax(attention_score / (Key.shape[-1] ** 0.5), dim = -1)
```

```python
context_vector = attention_weight @ inputs
print(context_vector)
```

```python
import torch.nn as nn
class SelfAttention_v1(nn.Module):
  def __init__(self, din, dout):
    super().__init__()
    self.wq = nn.Parameter(torch.rand(din, dout), requires_grad=False)
    self.wk = nn.Parameter(torch.rand(din, dout), requires_grad=False)
    self.wv = nn.Parameter(torch.rand(din, dout), requires_grad=False)
  def forward(self, x):
    query = x @ self.wq
    key = x @ self.wk
    value = x @ self.wv
    attention_scores = query @ key.T
    attention_weights = torch.softmax(attention_scores / (key.shape[-1] ** 0.5) , dim = -1)
    context_vector = attention_weights @ value
    return context_vector
```

```python
torch.manual_seed(123)
att = SelfAttention_v1(din, dout)
print(att(inputs))
```

```python
class SelfAttention_v2(nn.Module):
  def __init__(self, din, dout, qkv_bias=False):
    super().__init__()
    self.Wq = nn.Linear(din, dout, bias = qkv_bias)
    self.Wk = nn.Linear(din, dout, bias = qkv_bias)
    self.Wv = nn.Linear(din, dout, bias = qkv_bias)
  def forward(self, x):
    query = x @ Wq
    key = x @ Wk
    value = x @ Wv
    attention_scores = query @ key.T
    attention_weights = torch.softmax(attention_scores / (key.shape[-1] ** 0.5) , dim = -1)
    context_vector = attention_weights @ value
    return context_vector
```

```python
att = SelfAttention_v2(din, dout)
print(att(inputs))
```
