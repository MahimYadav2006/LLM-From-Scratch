```python
# Reading Files
with open("/content/text.txt", "r", encoding="utf-8") as file:
  content = file.read()
  print(content[:20])
```

```python
# Finding all non-alphanumeric characters
s = set()
for i in content:
  if(i.isalnum() == False):
    s.add(i)
for i in s:
  print(i)
```

```python
# Matching pattern and breaking into words
import re
regex_pattern = r"""([=,$\/%@.:?\];~!'*#\[_\-\(\)"]|\s+)"""
all_words = re.split(regex_pattern, content)
print(all_words[:50])
```

```python
sorted_words = sorted(set(all_words))
sorted_words.append("<|unk|>")
sorted_words.append("<|endOfText|>")

vocab = {word:i for i,word in enumerate(sorted_words)}
```

```python
class TokenizerV1:
  def __init__(self, vocab):
    self.text_to_int = vocab;
    self.int_to_text = {i: word for word, i in vocab.items()}
  def encode(self, text):
    tokenized_text = re.split(regex_pattern, text)
    token_words = [item if item in vocab else "<|unk|>" for item in tokenized_text]
    # print(token_words)
    encoded = [self.text_to_int[token] for token in token_words]
    return encoded
  def decode(self, encoded):
    decoded = [self.int_to_text[token_id] for token_id in encoded]
    ans = ""
    for token_word in decoded:
      ans += token_word
    return ans
```

```python
tokenizer = TokenizerV1(vocab)
sentence1 = "Hi, this is sentence 1."
sentence2 = "Hi, this is sentence 2"
input_sentence = sentence1 + "<|endOfText|> " + sentence2

print(tokenizer.encode(input_sentence))
print(tokenizer.decode(tokenizer.encode(input_sentence)))
```
